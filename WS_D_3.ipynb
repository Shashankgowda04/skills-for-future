```python
import pandas as pd
data = {
    "Energy Source": ["Solar", "Wind", "Hydropower", "Geothermal", "Biomass", "Nuclear"],
    "Energy Consumption (MWh)": [1200, np.nan, 2900, np.nan, 2500, 3200],
    "Cost (Million $)": [200, 400, np.nan, 150, 250, np.nan]
}
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.isnull().sum()
```

```python
cleaned_df=df.dropna()
print(cleaned_df)
```

```python
df['Energy Consumption (MWh)'] = df['Energy Consumption (MWh)'].fillna(df['Energy Consumption (MWh)'].mean())
df['Cost (Million $)'] = df['Cost (Million $)'].fillna(df['Cost (Million $)'].mean())
print(df)
```

```python
df['cost per MW']=df['Energy Consumption (MWh)']/df['Cost (Million $)']
df
```

# Data preprocessing

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df['Energy Consumption (MWh)']=scaler.fit_transform(df[['Energy Consumption (MWh)']])
df['Cost (Million $)']=scaler.fit_transform(df[['Cost (Million $)']])
df
```

```python
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
df[['Energy Consumption (MWh)','Cost (Million $)']]=std_scaler.fit_transform(df[['Energy Consumption (MWh)','Cost (Million $)']])
df
```

```python
energy_encoded_df=pd.get_dummies(df,columns=['Energy Source'])
energy_encoded_df
```

```python
months=['Jan','Feb','Mar','Apr','May','Jun']
energy_consumption=[100,340,290,440,650,790]
```

```python
import matplotlib.pyplot as plt
plt.plot(months,energy_consumption ,marker='o',linestyle='dashed',color='blue')
plt.xlabel(months)
plt.ylabel(energy_consumption)
plt.title('Energy consumption')
plt.show()
```

```python
energy_sources=['Solar','Wind','Hydropower','Geothermal','Biomass']
energy_consumption=[1200,3400,2900,1800,2500]
plt.bar(energy_sources,energy_consumption,color="#9B111E")
plt.xlabel('Energy Source')
plt.ylabel('Energy Consumption (MWh)')
plt.title('Energy Consumption vs Source')
plt.show()
```

```python
plt.pie(energy_consumption,labels=energy_sources,autopct='%1.1f%%',colors=['#9B111E','#50C878','#000080','gold','silver','ivory','cream'])
plt.title('Energy Consumption vs Source')
plt.show()
```
