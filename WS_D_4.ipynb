polynomial regression

```python
#polynomial regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
#load the dataset
df=pd.read_csv('sustainability_data.csv')
df.tail()
```

```python
df.columns
```

```python
#feature selection
x=df[['Year','Energy_Consumption','Renewable_Percentage','GDP']]
y=df['CO2_Emissions']
```

```python
#split the data
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.8,random_state=42)
```

```python
#define the degree of polynomial
poly=PolynomialFeatures(degree=2)
x_train_poly=poly.fit_transform(x_train)
x_test_poly=poly.transform(x_test)
```

```python
#initialize and train the model
poly_model=LinearRegression()
poly_model.fit(x_train_poly,y_train)
```

```python
#prediction
poly_y_pred=poly_model.predict(x_test_poly)
poly_y_pred
```

```python
#evaluate
mse=mean_squared_error(y_test,poly_y_pred)
r2=r2_score(y_test,poly_y_pred)
print(f'Mean Squared Error is {mse}' )
print(f'R-squared Score is {r2}')
```

**classification**

```python
#import libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load the dataset
df=pd.read_csv('green_tech_data.csv')
df.head(10)
```

```python
#missing values
df.isnull().sum()
```

```python
#number of rows and columns
df.shape
```

```python
#column names
df.columns
```

```python
#Feature Selection
x=df[['carbon_emissions','energy_output', 'renewability_index',
       'cost_efficiency']]
y=df['sustainability']
```

```python
#split the data
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
#initialize and train
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
#train
model.fit(x_train,y_train)
```

```python
#prediction
y_pred=model.predict(x_test)
y_pred
```

```python
#metrics
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy Score is {accuracy}')
```

```python
#CM
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
import seaborn as sns
sns.heatmap(cm,annot=True,cmap='coolwarm',xticklabels=['not sustainable','sustainable'],yticklabels=['not sustainable','sustainable'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```

```python
cr=classification_report(y_test,y_pred,target_names=['not sustainable','sustainable'])
print(cr)
```

```python
#feature importance
co=pd.DataFrame(model.coef_.T,index=x.columns,columns=['coefficient'])
print(co)
```

```python
#save model
import joblib
joblib.dump(model,'lr_model.pkl')
```

```python
#import necessary libraries
import pandas as pd
import numpy as np
import seaborn as sns

```

```python
#load dataset
df=pd.read_csv('agricultural sustainability.csv')
df.head()
```

```python
#columns
df.columns
```

```python
#feature selection
x=df[['soil_health', 'crop_yield', 'water_usage', 'carbon_footprint',
       'fertilizer_use']]
y=df['sustainability']
print(x.shape)
print(y.shape)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
#initailize the model on training
from sklearn.ensemble import RandomForestClassifier
model=RandomForestClassifier()
model.fit(x_train,y_train)
```

```python
#predict
y_pred=model.predict(x_test)
print("The Predicted Values:")
print(y_pred)
```

```python
#metrics

accuracy=accuracy_score(y_test,y_pred)
print(f'Accuracy Score is {accuracy}')
```

```python
#confusion matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
#Classification report
cr=classification_report(y_test,y_pred,target_names=['Not sustainable','sustainable'])
print(cr)
```

```python
#lets save the model
import joblib
joblib.dump(model,'Random_forest_model.pkl')
```
